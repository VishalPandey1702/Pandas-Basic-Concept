```python
# Top Batsman On The Basis Of Runs
delivries = pd.read_csv('deliveries.csv')
delivries.head(2)
runs = delivries.groupby('batsman')
runs['batsman_runs'].sum().sort_values().tail(5) # top 5 batsman on the basis of runs
```

```python
# Using IS IN Function
# Q--> Best batsman on the basis of the strike rate of last overs (16-20)
delivry = pd.read_csv('deliveries.csv')
delivry
mask =delivry['over']>15 # It feth the rows of more than 15 over
death_over = delivry[mask]
death_over
all_batsman = death_over.groupby('batsman')['batsman_runs'].count() #It fetch those which are play more than 200 ball in ipl
x = all_batsman>200
batsman_list = all_batsman[x].index.tolist()
mask = delivry['batsman'].isin(batsman_list) # It return those which are in batsman list
final = delivry[mask]
run = final.groupby('batsman')['batsman_runs'].sum()
ball = final.groupby('batsman')['batsman_runs'].count()
strick_rate = (run/ball)*100
strick_rate.sort_values(ascending=False).head(5)
```

```python
# DAY 3
# Merg Function-->It Is Use To Merg The Two Data Of The Given Data 
# Q --> Orange cap Holder Of All Season
import pandas as pd
delivry = pd.read_csv('deliveries.csv')
match = pd.read_csv("matches - matches.csv")
match
delivry
```

```python
new = delivry.merge(match,left_on='match_id',right_on='id') # It merge the two dta items on the basis of left on and ight on
```

```python
orange_group = new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first') # reset index convert the file into the data form
orange_group.sort_values('season')
```
